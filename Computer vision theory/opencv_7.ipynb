Operaciones aritméticas con imágenes

```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la dirección para que no se confunda en la ruta 
img1 = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\perro.jpg')
img1=cv2.resize(img1,(500,500))

img2 = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\arbol.jpg')
img2=cv2.resize(img2,(500,500))

def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()
```

```python
resultadoSuma=cv2.add(img1,img2)#realizamos la suma de imágenes
```

```python
mostrar_imagen(resultadoSuma)
```

```python
resultadoResta=cv2.subtract(img2,img1)#realizamos la resta
mostrar_imagen(resultadoResta)
```

```python
resultadomultiplicacion=cv2.multiply(img1,img2)
mostrar_imagen(resultadomultiplicacion)
```

```python
resultadodivision=cv2.divide(img1,img2)
mostrar_imagen(resultadodivision)
```
