Segundo tema: Escalado de colores

```python
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\arbol.jpg')
#grabaremos las imágenes según su escala de color
B,G,R=cv2.split(img)
#unimos los 3 canales para tener una sola imagen
union_imagen=cv2.merge((B,G,R))

#creamos una función para mostrar la imagen de forma más sencilla
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()
    
#se muestran las imagenes   
mostrar_imagen(B)
mostrar_imagen(G)
mostrar_imagen(R)
mostrar_imagen(union_imagen)
```

```python
#espacio de color HSV, es mayormente utilizado para la detección de objetos mediante color
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\arbol.jpg')

hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

H,S,V=cv2.split(img)


mostrar_imagen(H)
mostrar_imagen(S)
mostrar_imagen(V)
mostrar_imagen(hsv)
```
