```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la dirección para que no se confunda en la ruta 
img1 = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\casa.jpg')
img1=cv2.resize(img1,(500,500))
img2 = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\casa_vacia.jpg')
img2=cv2.resize(img2,(500,500))
```

```python
resultadoresta=cv2.absdiff(img2,img1)#realizamos resta de imágenes donde el resutlado es el valor absoluto de la diferencia
```

```python
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

mostrar_imagen(img1)
mostrar_imagen(img2)
mostrar_imagen(resultadoresta)
```
