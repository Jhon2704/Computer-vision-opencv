```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la direccin para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\perro.jpg')

# Verificar si la imagen se carg correctamente
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("La imagen se carg correctamente.")

    
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

img2=cv2.resize(img,(100,100))#redimemsionamos la imagen para efectos prcticos
alto,ancho,canales=img2.shape
mostrar_imagen(img2)
```

```python
angle=90
center_image=(ancho // 2,alto // 2) #para tener el centro del giro desde el centro de la imagen
scale=1

M=cv2.getRotationMatrix2D(center_image,angle,scale)#comando para generar la matriz de rotacin 
print(M)

```

```python
imgT=np.zeros((alto+1,ancho+1,canales),np.uint8)#generamos una imagen oscura, el 1 lo ponemos ...

for i in range(alto):
    for j in range(ancho):
        px=np.array([j,i,1],np.uint8)
        dot=np.dot(M,px)
        x=int((dot[0]))
        y=int((dot[1]))
        imgT[y,x]=img2[i,j]


mostrar_imagen(imgT)
```

```python
#Utilizamos otra funcin de opnecv para hacer la rotacin

#leemos la imagen, se necesita poner la r delante de la direccin para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\perro.jpg')

# Verificar si la imagen se carg correctamente
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("La imagen se carg correctamente.")

    
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

img2=cv2.resize(img,(100,100))#redimemsionamos la imagen para efectos prcticos
alto,ancho,canales=img2.shape
mostrar_imagen(img2)

angle=110
center_image=(ancho // 2,alto // 2) #para tener el centro del giro desde el centro de la imagen
scale=1

M=cv2.getRotationMatrix2D(center_image,angle,scale)#comando para generar la matriz de rotacin 
print(M)



imgT=cv2.warpAffine(img2,M,(ancho,alto)) #con este comando de opencv podemos hacer la rotacin automticamente

mostrar_imagen(imgT)
```

```python
#vamos a explicar otra forma de rotar una imagen un poco ms sencillo 
import imutils

#leemos la imagen, se necesita poner la r delante de la direccin para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\perro.jpg')

# Verificar si la imagen se carg correctamente
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("La imagen se carg correctamente.")

    
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

img2=cv2.resize(img,(100,100))#redimemsionamos la imagen para efectos prcticos
alto,ancho,canales=img2.shape
mostrar_imagen(img2)

angle=110
center_image=(ancho // 2,alto // 2) #para tener el centro del giro desde el centro de la imagen

imgT=imutils.rotate_bound(img2,angle) #comando de imutils para girar la imagen de forma automtica con solo el angulo de la imagen en centro de eje cambia

mostrar_imagen(imgT)
```
