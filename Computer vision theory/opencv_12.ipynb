```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la dirección para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\foto antigua an.jpg',0)# en escala de grises
```

```python
equalizado=cv2.equalizeHist(img)#realizamos la ecualización del histograma
```

```python
res=np.hstack((img,equalizado))#concatenamos la imagen de entrada con la ecualizada
```

```python
cv2.namedWindow("resultado",cv2.WINDOW_NORMAL)#Construimos nuestra ventana donde se mostrara la imagen
cv2.imshow("resultado",res)#mostramos la imagen concatenada
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
hist,bins=np.histogram(img.flatten(),256,[0,256])
plt.hist(img.flatten(),256,[0,256],color='b')
plt.xlim([0,256])
plt.legend('H',loc='upper left')
plt.show()
```

```python
hist,bins=np.histogram(equalizado.flatten(),256,[0,256])
plt.hist(equalizado.flatten(),256,[0,256],color='b')
plt.xlim([0,256])
plt.legend('H',loc='upper left')
plt.show()
```

Ecualización adaptativa 

```python
#leemos la imagen, se necesita poner la r delante de la dirección para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\foto antigua an.jpg',0)# en escala de grises
```

```python
clahe=cv2.createCLAHE(clipLimit=10.0,tileGridSize=(8,8))#creamos la configuración para el algoritmo CLAHE con un umbral de 1.0 y un kernel de 8x8
cl1=clahe.apply(img)#aplicamos el metodo a la imagen de entrada
res=np.hstack((img,cl1))#concatenamos la imagen de entrada con la imagen ecualizada
#generamos una ventana para mostrarlo
cv2.namedWindow("resultado",cv2.WINDOW_NORMAL)#Construimos nuestra ventana donde se mostrara la imagen
cv2.imshow("resultado",res)#mostramos la imagen concatenada
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
hist,bins=np.histogram(img.flatten(),256,[0,256])
plt.hist(img.flatten(),256,[0,256],color='b')
plt.xlim([0,256])
plt.legend('H',loc='upper left')
plt.show()
hist,bins=np.histogram(cl1.flatten(),256,[0,256])
plt.hist(cl1.flatten(),256,[0,256],color='b')
plt.xlim([0,256])
plt.legend('H',loc='upper left')
plt.show()
```
