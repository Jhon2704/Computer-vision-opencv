```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la direccin para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\perro.jpg')

# Verificar si la imagen se carg correctamente
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("La imagen se carg correctamente.")

    
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

img2=cv2.resize(img,(50,50))#redimemsionamos la imagen para efectos prcticos
alto,ancho,canales=img2.shape#obtenemos las dimensiones de la imagen
mostrar_imagen(img2)
```

```python
#vamos a definir un factor de escala
escala=1
tMatrix=np.array([[escala,0,0],[0,escala,0],[0,0,1]])#generamos la matriz de transformacin

img2=np.zeros((alto*escala,ancho*escala,3),np.uint8)#generamos la matriz que recibe la imagen escalada 

for i in range(alto):
    for j in range(ancho):
        pixel=[i,j,0]#obtenemos el valor del pixel
        vectorPos=np.array([j,i,1])#generamos nuestro vector posicion
        resultPP=np.dot(tMatrix,vectorPos)#resolvemos el producto escalar de la matriz de transformacin con el vector posicin
        #print(resultPP)
        x=resultPP[0]
        y=resultPP[1]
        img2[y,x]=pixel

mostrar_imagen(img)
mostrar_imagen(img2)
        
        
        
        
```
