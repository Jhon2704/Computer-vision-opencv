Cuarto tema : Trabajando con pixels de la imagen

```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generamos una imagen oscura mediante numpy
imagenOscura=np.zeros((100,100,3),np.uint8) #generamos la matriz de 3 dimensiones y la llenamos con ceros

#leemos el valor de algunos de los pixeles de la imagen oscura
pixel=imagenOscura[97,97]#valor aleatorio

print(pixel)#mostramos el valor por pantalla

imagenOscura[97,97]=[255,255,255]#cambiamos el valor del pixel de la posición 97,97 de la matriz y lo cambiamos de negro a blanco

print(pixel) #el valor del pixel en este momento debe de ser el valor del blanco

alto,largo,canales=imagenOscura.shape#se utiliza para verificar el alto ancho y el número de canales de la imagen creada

print(alto,largo,canales)
```

```python
#Utilizamos la función para mostrar la imagen
def mostrar_imagen(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Oculta los ejes
    plt.show()

mostrar_imagen(imagenOscura) 
#se puede compronar al final de la imagen que hay un pixel blanco 
```

```python
#para recorrer las matrices de una imagen es muy habitual utilizar un bucle anidado
for i in range(largo):
    for j in range(alto):
        print(imagenOscura[i,j])
```

```python
#Modificamos valores de cada pixel
for i in range(largo):
    for j in range(alto):
        pixel=imagenOscura[i,j]
        if pixel[0]==0 and pixel[1]==0 and pixel[2]==0: #como es una matriz de 3 dimensiones tenemos que cada valor de pixel equivale a una columna
            imagenOscura[i,j]=[0,0,255]
            
mostrar_imagen(imagenOscura)
```

Septimo tema: Select ROI

```python
#leemos la imagen, se necesita poner la r delante de la dirección para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\candados.jpg')

# Verificar si la imagen se cargó correctamente
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    print("La imagen se cargó correctamente.")
```

```python
#obtenemos las dimensiones
alto,largo,canales=img.shape
```

```python
#seleccionamos una región de interés
roi=cv2.selectROI("ROI1",img)#con este comando podemos seleccionar una sección de la imagen que querremos a posterior modificar, interacción manual en la imagen
print(roi)
```

```python
candado1=img[int(roi[1]):int(roi[1]+roi[3]),int(roi[0]):int(roi[0]+roi[2])] #son las dimensiones de la roi en la imagen
alto1,largo1,_=candado1.shape #mostrar las dimensiones de la ROI 1
```

```python
#seleccionamos una región de interés
roi2=cv2.selectROI("ROI2",img)#con este comando podemos seleccionar una sección de la imagen que querremos a posterior modificar, interacción manual en la imagen
print(roi2)
```

```python
candado2=img[int(roi2[1]):int(roi2[1]+roi2[3]),int(roi2[0]):int(roi2[0]+roi2[2])]#son las dimensiones de la roi en la imagen
alto2,largo2,_=candado2.shape#mostrar las dimensiones de la ROI 2
```

```python
#redimensionamos las ROI seleccionadas para que coincidan
newCandado1=cv2.resize(candado1,(largo2,alto2)) 
newCandado2=cv2.resize(candado2,(largo1,alto1))
```

```python
#Hacemos la combinación de la imagenes para conseguir una imagen compuesta con las dos ROI'S
img[int(roi[1]):int(roi[1]+roi[3]),int(roi[0]):int(roi[0]+roi[2])]=newCandado2
img[int(roi2[1]):int(roi2[1]+roi2[3]),int(roi2[0]):int(roi2[0]+roi2[2])]=newCandado1

mostrar_imagen(img)
```
