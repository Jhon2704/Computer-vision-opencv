Matriz de traslaci贸n




```python
tx=10 #definimos la traslaci贸n en x de 10 px
ty=2 #definimos la traslaci贸n en y de 2 px

alto,ancho,canales=img2.shape#obtenemos las dimensiones de la imagen

mT= np.array(([1,0,tx],[0,1,ty],[0,0,1]),np.uint8)#la matriz de transformaci贸n

imgT=np.zeros((alto+2,ancho+10,3),np.uint8)
```

```python
for i in range(alto):#recorremos las filas de la matriz de la imagen
    for j in range(ancho):#recorremos las columnas de la matriz de la imagen
        px=np.array(([j,i,1]),np.uint8)#vector de posici贸n
        dot=np.dot(mT,px)#realizamos el producto punto entre la matriz de transformaci贸n y el vector de posici贸n
        x=dot[0]#extraemos el valor para x
        y=dot[1]#extraemos el valor para y
        imgT[y,x]=img2[i,j]#remplazamos los valores de los pixeles 
```

```python
mostrar_imagen(imgT)
```
