```python
#importamos la libreria de opencv y matplotlib para poder visualizar la imagen correctamente
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#leemos la imagen, se necesita poner la r delante de la direcci√≥n para que no se confunda en la ruta 
img = cv2.imread(r'C:\Users\joano\Desktop\vision artificial juan olivan\imagenes\lena.jpg',0)#imagen es escala de grises
```

```python

mask=np.zeros(img.shape[:2],np.uint8)#creamos una imagen en negro en unidades de 8 bits
mask[0:100,0:100]=255#cambiamos los valores de la matriz 0-100 a blanco
masked_img=cv2.bitwise_and(img,img,mask=mask)##aplicamos una compuerta and bit abit entre img img,pero solo en la parte de la m√°scara con pixeles blancos

#calculamos los histogramas
histCompleto=cv2.calcHist([img],[0],None,[256],[0,255])
histMask=cv2.calcHist([img],[0],mask,[256],[0,255])

#graficamos los resultados
plt.subplot(221),plt.imshow(img,"gray")
plt.subplot(222),plt.imshow(mask,"gray")
plt.subplot(223),plt.imshow(masked_img,"gray")
plt.subplot(224),plt.plot(histCompleto),plt.plot(histMask)
plt.xlim([0,255])
plt.show()
```
